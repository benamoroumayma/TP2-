# Partie 1

```python
import numpy as np
import cv2
from sklearn.cluster import KMeans
from sklearn.cluster import MiniBatchKMeans
import skimage.io
import matplotlib.pyplot as plt

image = cv2.imread('imgorginal.jpg')
(h1, w1) = image.shape[:2]

image = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

image = image.reshape((image.shape[0] * image.shape[1], 3))

clt = KMeans(n_clusters = 8)

labels = clt.fit_predict(image)
quant = clt.cluster_centers_.astype("uint8")[labels]

#reshape the feature vectors to images
quant = quant.reshape((h1, w1, 3))
image = image.reshape((h1, w1, 3))

# convert from L*a*b* to RGB
quant = cv2.cvtColor(quant, cv2.COLOR_LAB2BGR)
image = cv2.cvtColor(image, cv2.COLOR_LAB2BGR)

cv2.imwrite('normal6.jpg', quant)
plt.imshow(image)

cv2.imread('imgorginal.jpg')
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Partie 2

#### *On va ajouter un bruit ("gaussian") a notre image puis on faire l'alogithem de KMeans et on voir le rsultat*

```python
import skimage.io
import matplotlib.pyplot as plt
img_path="imgorginal.jpg"
img = skimage.io.imread(img_path)/255.0

def plotnoise(img, mode, r, c, i):
    plt.subplot(r,c,i)
    if mode is not None:
        gimg = skimage.util.random_noise(img, mode=mode)
        plt.imshow(gimg)
        plt.imsave("imagnoise.jpg",gimg)
        
    else:
        plt.imshow(img)
    plt.title(mode)
    plt.axis("off")

plt.figure(figsize=(18,24))

r=4
c=2
plotnoise(img, "gaussian", r,c,1)
#plotnoise(img, "localvar", r,c,2)
#plotnoise(img, "poisson", r,c,3)
#plotnoise(img, "salt", r,c,4)
#plotnoise(img, "pepper", r,c,5)
#plotnoise(img, "s&p", r,c,6)
#plotnoise(img, "speckle", r,c,7)
#plotnoise(img, None, r,c,8)

plt.show()
```

```python
image = cv2.imread('imagnoise.jpg')
(h1, w1) = image.shape[:2]

image = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

image = image.reshape((image.shape[0] * image.shape[1], 3))

clt = KMeans(n_clusters = 8)

labels = clt.fit_predict(image)
quant = clt.cluster_centers_.astype("uint8")[labels]

#reshape the feature vectors to images
quant = quant.reshape((h1, w1, 3))
image = image.reshape((h1, w1, 3))

# convert from L*a*b* to RGB
quant = cv2.cvtColor(quant, cv2.COLOR_LAB2BGR)
image = cv2.cvtColor(image, cv2.COLOR_LAB2BGR)

cv2.imwrite('imagenoise+clust.jpg', quant)
plt.imshow(image)

cv2.imread('imagnoise.jpg')
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Partie 3
### *On va faire un algo qui elimine le bruit puis on va applique l'alorithem de clustering et on va voir le rsultat*

```python
from skimage import io,img_as_float
from scipy import ndimage as nd
import matplotlib.pyplot as plt
import numpy as np



#img = skimage.io.imread("imagnoise.jpg")
#gaussian_img =nd.gaussian_filter(img, sigma=4)
#plt.imsave("imag.jpg",gaussian_img)



from skimage.restoration import denoise_nl_means, estimate_sigma

img = img_as_float(io.imread("imagnoise.jpg"))
sigma_est =np.mean(estimate_sigma(img,multichannel=True))
mm=denoise_nl_means(img,h=1.15 * sigma_est, fast_mode=True, patch_size=5,patch_distance=3, multichannel=True)
plt.imsave("mm.jpg",mm)

```

```python
image = cv2.imread('mm.jpg')
(h1, w1) = image.shape[:2]

image = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

image = image.reshape((image.shape[0] * image.shape[1], 3))

clt = KMeans(n_clusters = 8)

labels = clt.fit_predict(image)
quant = clt.cluster_centers_.astype("uint8")[labels]

#reshape the feature vectors to images
quant = quant.reshape((h1, w1, 3))
image = image.reshape((h1, w1, 3))

# convert from L*a*b* to RGB
quant = cv2.cvtColor(quant, cv2.COLOR_LAB2BGR)
image = cv2.cvtColor(image, cv2.COLOR_LAB2BGR)

cv2.imwrite('mm+clust.jpg', quant)
plt.imshow(image)

cv2.imread('mm.jpg')
cv2.waitKey(0)
cv2.destroyAllWindows()
```
